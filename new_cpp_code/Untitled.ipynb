```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import KDTree
import math
```

```python
N = 1000
r = np.random.rand(N)
```

```python
r = r.reshape(N//2,2)

x, y = r[:,0], r[:, 1]
```

```python
plt.scatter(x, y)
```

```python
tree = KDTree(r)
```

```python
point_ind = np.random.randint(0, len(r))
ind = tree.query_ball_point(r[point_ind], r=0.2)


fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot()
ax.plot(x, y, 'b.')
ax.plot(x[ind], y[ind], 'r.')
ax.plot(x[point_ind], y[point_ind], 'k*')
ax.set_xlim(0,1)
ax.set_ylim(0,1)


def drawCircle(point, radius, ax):
    '''Draw a quarter of a circle.'''
    xpoints, ypoints = point[0] + radius * np.cos(np.radians(np.arange(360))), point[1] + radius * np.sin(np.radians(np.arange(360)))
    if ax:
        ax.plot(xpoints, ypoints, 'c')
    else:
        plt.plot(xpoints, ypoints, 'c')

drawCircle([x[point_ind], y[point_ind]], 0.2, ax)
```
